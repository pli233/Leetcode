```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        
        left= 0
        right = len(numbers)-1
        
        while right < len(numbers) and left < right:
            print("left: "+str(left))
            print("right: "+ str(right))
            cursum = numbers[left]+numbers[right]
            if cursum == target:
                return [left+1, right+1]
            elif cursum < target:
                left += 1
            else:
                right -= 1
```
