```python
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        
        #解题思路：用一个慢指针和一个快指针来update 数组，只要fast指向的内容是符合要求的
        #update到slow指针的位置
        slow = 0
        for fast in range(len(nums)):
            if nums[fast] != val:
                nums[slow] = nums[fast]
                slow +=1
        return slow
```
