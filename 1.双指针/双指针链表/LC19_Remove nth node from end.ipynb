```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def removeNthFromEnd(self, head, n):
        """
        :type head: Optional[ListNode]
        :type n: int
        :rtype: Optional[ListNode]
        """
        #Handle enpty linked list
        if head is None:
            return None

        todelete = self.getNthFromEnd(head, n)
        if head == todelete:
            return head.next
            
        slow = head
        fast = head.next 

        while slow is not None:
            if fast == todelete:
                slow.next = fast.next
                return head
            fast = fast.next
            slow = slow.next

    def getNthFromEnd(self, head, n):
    
        fast = head
        for i in range(n):
            if fast is None:
                return None
            fast = fast.next

        
        slow = head
        while fast is not None:
            fast = fast.next
            slow = slow.next
        return slow
```
