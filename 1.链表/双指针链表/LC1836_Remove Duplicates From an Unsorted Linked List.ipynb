```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicatesUnsorted(self, head: ListNode) -> ListNode:
        if head is None:
            return None

        # 
        freq = {}
        p = head
        while p is not None:
            freq[p.val] = freq.get(p.val, 0) + 1
            p = p.next
        
        dummyhead = ListNode(-1)

        cur = dummyhead
        prev = head

        while prev is not None:
            if freq[prev.val] == 1:  # 
                cur.next = prev  # 
                cur = cur.next  # 
            prev = prev.next  # 
        cur.next = None
        return dummyhead.next
```
