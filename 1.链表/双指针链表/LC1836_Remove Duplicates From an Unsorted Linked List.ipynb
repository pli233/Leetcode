```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicatesUnsorted(self, head: ListNode) -> ListNode:
        if head is None:
            return None

        # 第一次遍历：统计每个值的出现次数
        freq = {}
        p = head
        while p is not None:
            freq[p.val] = freq.get(p.val, 0) + 1
            p = p.next
        
        dummyhead = ListNode(-1)

        cur = dummyhead
        prev = head

        while prev is not None:
            if freq[prev.val] == 1:  # 如果节点值只出现一次
                cur.next = prev  # 将节点连接到结果链表
                cur = cur.next  # 更新结果链表尾部
            prev = prev.next  # 移动到下一个节点
        cur.next = None
        return dummyhead.next
```
