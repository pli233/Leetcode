```python
import heapq

class Solution:
    def kSmallestPairs(self, nums1: List[int], nums2: List[int], k: int) -> List[List[int]]:

        #Initialize the sum of each element in nums1 and the first element in nums2

        pq = []
        for i in range(len(nums1)):
            e1 = nums1[i]
            e2 = nums2[0]
            cursum = e1+e2

            heapq.heappush(pq, (cursum, i, 0))
        
        res = []
        while pq and k>0:
            cursum, index1, index2 = heapq.heappop(pq)
            res.append([nums1[index1], nums2[index2]])
            k = k-1

            #index2rangeupdate index1,index1
            #pq
            next_index2 = index2+1
            if  next_index2 < len(nums2):
                nextsum = nums1[index1] + nums2[next_index2]
                heapq.heappush(pq, (nextsum, index1, next_index2))

        return res

        
```
