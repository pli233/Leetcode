```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        #Base case 1: empty linked list
        if head is None:
            return None
        
        #Base case 2: one element linked list
        if head.next is None:
            return head

        dummyhead = ListNode(-1)
        p = dummyhead
        slow = head
        fast = head.next
        distance = 1

        while fast is not None:
            print("slow: "+str(slow.val), end = " ")
            print("fast: "+str(fast.val), end = " ")
            print("distance: "+str(distance))

            #1.slowvaluefastunique valappend
            # 1 2 3
            #   s f
            if slow.val != fast.val and distance == 1:

                #record slow
                p.next = slow
                p = p.next

                #do update
                slow = fast
                fast = fast.next
                distance = 1

            #2.slowvaluefastfastvalupdate slowref
            #unique val
            # 1 2 2 2 3
            #   s     f
            elif slow.val != fast.val and distance != 1:   
                slow = fast
                fast = fast.next
                distance = 1
            
            #2.slowvaluefastupdate fastnextref
            elif slow.val == fast.val:
                fast = fast.next
                distance += 1
        
        # Check the last element
        if distance == 1:
            p.next = slow
            p = p.next

        # Terminate the result list
        p.next = None

        return dummyhead.next
        
```

```python
class Solution:
    def deleteDuplicates(self, head: ListNode) -> ListNode:
        # Dummy node to simplify edge cases
        dummy = ListNode(-1)
        dummy.next = head
        
        slow = dummy
        fast = head
        
        while fast is not None:
            # 
            if fast.next is not None and fast.val == fast.next.val:
                # 
                while fast.next is not None and fast.val == fast.next.val:
                    fast = fast.next
                #  slow.next
                slow.next = fast.next
            else:
                #  slow 
                slow = slow.next
            
            #  fast 
            fast = fast.next
        
        return dummy.next
```
