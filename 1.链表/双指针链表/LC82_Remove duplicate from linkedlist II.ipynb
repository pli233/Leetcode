```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        #Base case 1: empty linked list
        if head is None:
            return None
        
        #Base case 2: one element linked list
        if head.next is None:
            return head

        dummyhead = ListNode(-1)
        p = dummyhead
        slow = head
        fast = head.next
        distance = 1

        while fast is not None:
            print("slow: "+str(slow.val), end = " ")
            print("fast: "+str(fast.val), end = " ")
            print("distance: "+str(distance))

            #1.slow的value和fast不一样，且相邻，说明我们获得一个unique val，应该append
            # 1 2 3
            #   s f
            if slow.val != fast.val and distance == 1:

                #record slow
                p.next = slow
                p = p.next

                #do update
                slow = fast
                fast = fast.next
                distance = 1

            #2.slow的value和fast不一样，但不相邻，说明我们fast已经到达一个新的val，应该update slow的ref
            #但这不是unique val
            # 1 2 2 2 3
            #   s     f
            elif slow.val != fast.val and distance != 1:   
                slow = fast
                fast = fast.next
                distance = 1
            
            #2.slow的value和fast一样，我们update fast作为next的ref
            elif slow.val == fast.val:
                fast = fast.next
                distance += 1
        
        # Check the last element
        if distance == 1:
            p.next = slow
            p = p.next

        # Terminate the result list
        p.next = None

        return dummyhead.next
        
```

```python
class Solution:
    def deleteDuplicates(self, head: ListNode) -> ListNode:
        # Dummy node to simplify edge cases
        dummy = ListNode(-1)
        dummy.next = head
        
        slow = dummy
        fast = head
        
        while fast is not None:
            # 如果发现重复节点
            if fast.next is not None and fast.val == fast.next.val:
                # 跳过所有重复节点
                while fast.next is not None and fast.val == fast.next.val:
                    fast = fast.next
                # 跳过重复节点，更新 slow.next
                slow.next = fast.next
            else:
                # 不重复，移动 slow 指针
                slow = slow.next
            
            # 移动 fast 指针
            fast = fast.next
        
        return dummy.next
```
