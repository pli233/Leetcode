```python
from typing import List
class NumMatrix:
    # preSum[i][j]  matrix  [0, 0, i-1, j-1] 
    def __init__(self, matrix: List[List[int]]):
        m, n = len(matrix), len(matrix[0])
        if m == 0 or n == 0: return
        # 
        preSum = [[0 for _ in range(n + 1)] for _ in range(m + 1)]
        for i in range(1, m + 1):
            for j in range(1, n + 1):
                #  [0, 0, i, j] 
                preSum[i][j] = (
                    matrix[i - 1][j - 1]
                    + preSum[i - 1][j]
                    + preSum[i][j - 1]
                    - preSum[i - 1][j - 1]
                )

        self.preSum = preSum
        #  preSum 
        self.printPreSum()

    #  preSum 
    def printPreSum(self):
        for row in self.preSum:
            print(" ".join(f"{v:5}" for v in row))  #  5 

    #  [x1, y1, x2, y2] 
    def sumRegion(self, x1: int, y1: int, x2: int, y2: int) -> int:
        print(self.preSum[x2 + 1][y2 + 1])
        print(self.preSum[x1][y2+1])
        print(self.preSum[x2+1][y1])
        print(self.preSum[x1][y1])
        # 
        return (
            self.preSum[x2 + 1][y2 + 1]
            - self.preSum[x1][y2 + 1]
            - self.preSum[x2 + 1][y1]
            + self.preSum[x1][y1]
        )
```

```python
test = [[3,0,1,4,2],[5,6,3,2,1],[1,2,0,1,5],[4,1,0,1,7],[1,0,3,0,5]]
sol = NumMatrix(test)
```
