```python
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        n = len(nums)
        preSum = [0] * (n+1)

        for i in range(1, n+1):
            preSum[i] = preSum[i-1]+nums[i-1]
        
        for i in range(1, n+1):
            leftSum = preSum[i-1]
            rightSum = preSum[-1] - preSum[i]

            if leftSum == rightSum:
                return i-1
    
        return -1
```
