```python
class Solution:
    def checkSubarraySum(self, nums: List[int], k: int) -> bool:
        n = len(nums)
        preSum = [0] * (n+1)
        

        for i in range(1, n+1):
            preSum[i] = preSum[i-1] + nums[i-1]

        #map: remainder: idx1
        remainder_to_idx = {}

        for i in range(n+1):
            remainder = preSum[i] % k
            
            if remainder not in remainder_to_idx:
                remainder_to_idx[remainder] = i
            else:
                if i - remainder_to_idx[remainder] > 1:
                    return True
        return False
```
