```python
class Solution:
    def findMaxLength(self, nums: List[int]) -> int:
        
        n = len(nums)
        preSum = [0] * (n+1)
        for i in range(1, n+1):
            if nums[i-1] == 0:
                preSum[i] = preSum[i-1] + (-1)
            else:
                preSum[i] = preSum[i-1] + 1

        #only first index will be record
        #val: first_index_appeared
        val_index_map = {}
        res = 0
        for i in range(len(preSum)):
            #index
            if preSum[i] not in val_index_map:
                val_index_map[preSum[i]] = i
            else:
                #  0 
                # preSumindex
                res = max(res, i - val_index_map[preSum[i]])
        return res
```
