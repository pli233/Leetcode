```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        n = len(nums)
        # 
        preSum = [0] * (n + 1)
        # 
        count = {0: 1}
        #  k 
        res = 0

        #  nums 
        for i in range(1, n + 1):
            preSum[i] = preSum[i - 1] + nums[i - 1]
            #  need 
            #  nums[i-1]  k
            need = preSum[i] - k
            if need in count:
                res += count[need]
            # 
            if preSum[i] not in count:
                count[preSum[i]] = 1
            else:
                count[preSum[i]] = count[preSum[i]] + 1
        return res
```
