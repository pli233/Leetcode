```python
class Solution:
    def subarraysDivByK(self, nums: List[int], k: int) -> int:
        n = len(nums)
        preSum = [0] * (n+1)

        for i in range(1, n+1):
            preSum[i] = preSum[i-1] + nums[i-1]
        
        # {remainder1: 0, remainder2: 8}
        total_count = 0
        remainder_to_count = {}
        for i in range(n+1):
            remainder = preSum[i] % k
            if remainder not in remainder_to_count:
                remainder_to_count[remainder] = 1
            else:
                total_count += remainder_to_count[remainder]
                remainder_to_count[remainder]  += 1
        return total_count
```
