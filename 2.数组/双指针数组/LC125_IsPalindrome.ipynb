```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        
        regular_s = ""
        for c in s:
            if c.isalpha() or c.isnumeric():
                regular_s += c.lower()
        
        left = 0
        right = len(regular_s)-1

        while left < right:
            if regular_s[left] != regular_s[right]:
                return False
            left +=1
            right -= 1
        return True
        
```
