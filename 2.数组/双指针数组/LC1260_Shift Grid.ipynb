```python
class Solution:
    def shiftGrid(self, grid: List[List[int]], k: int) -> List[List[int]]:
        s = []
        n = len(grid)
        m = len(grid[0])

        for i in range(n):
            for j in range(m):
                s.append(grid[i][j])
        k = k % len(s)
        s2 = s[-k:] + s[:-k]
    
        for i in range(n):
            for j in range(m):
                grid[i][j] = s2.pop(0)
        return grid

        
```
