```python
class Solution:
    def diagonalSort(self, mat: List[List[int]]) -> List[List[int]]:
        
        n = len(mat)
        m = len(mat[0])
        diagonals = {}
        for i in range(n):
            for j in range(m):
                diagonalID = i-j
                if diagonalID not in diagonals:
                    diagonals[diagonalID] = [mat[i][j]]
                else:
                    diagonals[diagonalID].append(mat[i][j])
        
        for diagonalID, diagonalVals in diagonals.items():
            diagonalVals.sort()
        
        for i in range(n):
            for j in range(m):
                diagonalID = i-j
                mat[i][j] = diagonals[diagonalID].pop(0)
        return mat
        
```
