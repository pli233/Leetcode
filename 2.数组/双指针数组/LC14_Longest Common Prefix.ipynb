```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        n = len(strs)
        
        p = 0
        while p < len(strs[0]):
            temp = strs[0][p]
            for i in range(1, n):
                if p >= len(strs[i]):
                    return strs[0][:p]
                if strs[i][p] != temp:
                    return strs[0][:p]
            p += 1
        return strs[0][:p]
                    
```
