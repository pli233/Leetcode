```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """

        slow = 0
        fast = 0

        #modify values in place
        while fast < len(nums):
            if nums[fast] != 0:
                nums[slow] = nums[fast]
                slow +=1
            fast += 1
        
        #update remaining values to zero
        while slow < len(nums):
            nums[slow] = 0
            slow +=1
```
