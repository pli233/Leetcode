```python
class Solution:
    def sortTransformedArray(self, nums: List[int], a: int, b: int, c: int) -> List[int]:
        n = len(nums)
        left = 0
        right = n-1
        if a > 0:
            p = n-1
        else:
            p = 0
        res = [0] * n

        while left <= right:
            v_left = self.funcX(nums[left], a, b, c)
            v_right = self.funcX(nums[right], a, b, c)

            if a>0:
                if v_left >= v_right:
                    res[p] = v_left
                    left += 1
                else:
                    res[p] = v_right
                    right -= 1
                p -= 1
            else:
                if v_left >= v_right:
                    res[p] = v_right
                    right -= 1
                else:

                    res[p] = v_left
                    left += 1
                p += 1
        return res
            

    def funcX(self, x: int, a: int, b: int, c: int) -> int:
        return a*x**2 + b*x + c
```
