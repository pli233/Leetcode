```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        
        slow = 0
        count = 0
        prev = 0
        for fast in range(0, len(nums)):
            if prev == nums[fast]:
                count += 1
            else:
                prev = nums[fast]
                count = 1
            
            if count <= 2:
                nums[slow] = prev
                slow += 1
        return slow
```
