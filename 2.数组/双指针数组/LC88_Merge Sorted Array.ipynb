```python
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        # 定义指针
        p1 = m - 1  # nums1 中的有效元素指针
        p2 = n - 1  # nums2 中的元素指针
        p3 = m + n - 1  # nums1 的末尾指针

        # 从后往前合并
        while p2 >= 0:  # nums2 必须完全合并到 nums1 中
            if p1 >= 0 and nums1[p1] >= nums2[p2]:
                nums1[p3] = nums1[p1]
                p1 -= 1
            else:
                nums1[p3] = nums2[p2]
                p2 -= 1
            p3 -= 1
```
