```python
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        
        n = len(nums)
        left = 0
        right = n-1
        res = [0]*n
        p = n-1

        while left <= right:
            e_left = abs(nums[left])
            e_right = abs(nums[right])
            if e_left >= e_right:
                res[p] = e_left ** 2
                p = p-1
                left += 1
            else:
                res[p] = e_right ** 2
                p = p-1
                right -= 1
        return res
```
