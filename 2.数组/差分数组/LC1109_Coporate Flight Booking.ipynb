```python
class Diffarray:
    def __init__(self, nums):
        length = len(nums)
        data = [0] * length
        data[0] = nums[0]

        for i in range(1, length):
            data[i] = nums[i] - nums[i - 1]
        self.data = data
        self.length = length

    def get(self):
        res = [0] * self.length
        res[0] = self.data[0]

        for i in range(1, self.length):
            res[i] = res[i - 1] + self.data[i]
        return res

    def increase(self, i, j, v):
        self.data[i] += v
        if j + 1 < self.length:
            self.data[j + 1] -= v


class Solution:
    def corpFlightBookings(self, bookings: List[List[int]], n: int) -> List[int]:
        nums = [0] * n
        df_array = Diffarray(nums)

        for i, j, v in bookings:
            df_array.increase(i-1, j-1, v)
        return df_array.get()
```
