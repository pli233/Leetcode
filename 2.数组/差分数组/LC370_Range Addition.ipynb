```python
class DiffArray:
    def __init__(self, nums):
        assert len(nums) > 0

        length = len(nums)
        data = [0] * length
        data[0] = nums[0]
        for i in range(1, length):
            data[i] = nums[i] - nums[i - 1]

        self.data = data
        self.length = length

    def get(self):
        res = [0] * self.length
        res[0] = self.data[0]

        for i in range(1, self.length):
            res[i] = res[i - 1] + self.data[i]
        return res

    def increase(self, i, j, v):
        self.data[i] += v
        if j+1 < self.length:
            self.data[j+1] -= v

class Solution:
    def getModifiedArray(self, length: int, updates: List[List[int]]) -> List[int]:
        nums = [0] * length
        diffarray = DiffArray(nums)

        for i, j, v in updates:
            diffarray.increase(i, j, v)
        return diffarray.get()
```
