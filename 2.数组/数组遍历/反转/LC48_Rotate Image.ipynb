```python
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        
        n = len(matrix)
        #1. reverse the whole matrix 
        for i in range(n):
            for j in range(i+1, n):
                temp = matrix[i][j]
                matrix[i][j] = matrix[j][i]
                matrix[j][i] = temp

        #2. reverse the whole matrix 
        for i in range(n):
            for j in range(0, n // 2):
                pos2 = n-j-1
                temp = matrix[i][j]
                matrix[i][j] = matrix[i][pos2]
                matrix[i][pos2] = temp

```
