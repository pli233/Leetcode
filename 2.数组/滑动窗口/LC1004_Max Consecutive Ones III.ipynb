```python
class Solution:
    def longestOnes(self, nums: List[int], k: int) -> int:
        
        n = len(nums)
        left = 0
        right = 0
        count_0 = 0
        maxLength = 0

        while right < n:
            c = nums[right]
            right += 1
            
            if c == 0:
                count_0 += 1

            while left < right and count_0 > k:
                d = nums[left]
                left += 1

                if d == 0:
                    count_0 -= 1
            maxLength = max(maxLength, right-left)
        return maxLength



        
```
