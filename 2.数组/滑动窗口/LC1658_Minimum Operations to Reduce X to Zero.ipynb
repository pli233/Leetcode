```python
class Solution:
    def minOperations(self, nums: List[int], x: int) -> int:
        # sum(nums)-X
        target = sum(nums) - x
        if target < 0:
            return -1

        if target == 0:
            return len(nums)

        left = 0
        right = 0
        window_sum = 0
        longest = -1

        while right < len(nums):
            c = nums[right]
            right += 1
            window_sum += c

            while left < right and window_sum >= target:
                if window_sum == target:
                    longest = max(longest, right - left)
                d = nums[left]
                left += 1
                window_sum -= d
        if longest == -1:
            return -1
        else:
            return len(nums)-longest
```
