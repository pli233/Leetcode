```python
class Solution(object):
    def minSubArrayLen(self, target, nums):
        """
        :type target: int
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)      
        left = 0
        cur_sum = 0
        result = float('inf')

        for right in range(n):
            cur_sum += nums[right]
            
            while cur_sum >= target:
                result = min(result, right-left+1)
                cur_sum -= nums[left]
                left += 1
        if result != float('inf'):
            return result
        else:
            return 0

            
```
