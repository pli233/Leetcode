```python
class Solution:
    def containsNearbyDuplicate(self, nums: List[int], k: int) -> bool:
        left = 0 
        right = 0
        window = set()

        while right < len(nums):
            c = nums[right]
            right += 1

            if c in window:
                return True
            window.add(c)

            while left < right and right-left > k:
                d = nums[left]
                left += 1
                window.remove(d)

        return False
```
