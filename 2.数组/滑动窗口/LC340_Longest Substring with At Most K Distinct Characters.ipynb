```python
class Solution:
    def lengthOfLongestSubstringKDistinct(self, s: str, k: int) -> int:

        if k == 0:
            return 0

        left = 0
        right = 0
        counter = {}
        maxLength = 0

        while right < len(s):
            c = s[right]
            right += 1
            # counterright update+1
            counter[c] = counter.get(c, 0) + 1

            # windowdistinct valkshrink
            while left < right and len(counter) > k:
                d = s[left]
                left += 1
                if counter[d] == 1:
                    counter.pop(d)
                else:
                    counter[d] = counter[d] - 1
            maxLength = max(maxLength, right - left)
        return maxLength
```
