```python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        


        left = 0
        right = 0 
        counter = {}
        max_count = 0
        res = 0
        while right < len(s):

            c = s[right]
            right += 1
            counter[c] = counter.get(c,0) + 1

            max_count = max(max_count, counter[c])
        
            while left < right and (right-left-max_count) > k :
                d = s[left]
                left += 1
                counter[d] = counter[d] - 1

            res = max(res, right-left)
        return res
```
