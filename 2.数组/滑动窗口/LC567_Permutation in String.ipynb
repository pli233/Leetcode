```python
class Solution:
    def checkInclusion(self, t: str, s: str) -> bool:
        left = 0
        right = 0 
        valid = 0
        target = {}
        window = {}

        for c in t:
            target[c] = target.get(c,0) + 1
        
        while right < len(s):
            c = s[right]
            right += 1

            if c in target:
                window[c] = window.get(c, 0)+1
                if window[c] == target[c]:
                    valid += 1
                    if valid == len(target):
                        return True
            
            while right-left >= len(t):
                d = s[left]
                left += 1
                if d in target:
                    if window[d] == target[d]:
                        valid -= 1
                    window[d] -= 1
        return False
```
