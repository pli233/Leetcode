```python
class Solution:
    def numSubarrayProductLessThanK(self, nums: List[int], k: int) -> int:

        left = 0 
        right = 0
        count = 0
        curProduct = 1

        while right < len(nums):
            c = nums[right]
            right += 1
            curProduct *= c

            while left < right and curProduct >= k:
                d = nums[left]
                left += 1
                curProduct = curProduct // d
            count += right-left
        return count
        
```
