```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        
        window = {}

        left = 0
        right = 0
        res = 0
        while right < len(s):
            cur = s[right]
            right += 1
            if cur not in window:
                window[cur] = 1
            else:
                window[cur] = window[cur] + 1

            while left < right and window[cur] > 1:
                d = s[left]
                left += 1
                window[d] = window[d] -1
            res = max(res, right-left)
        return res
```

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        window = set()

        left = 0
        right = 0 
        longest_length = 0

        while right < len(s):
            c = s[right]
            right += 1
            #substring without repeating characters
            if c not in window:
                window.add(c)
            else:
                while left < right:
                    if s[left] == c:
                        left += 1
                        break
                    window.remove(s[left])
                    left += 1
            longest_length = max(longest_length, right-left)
        return longest_length
```
