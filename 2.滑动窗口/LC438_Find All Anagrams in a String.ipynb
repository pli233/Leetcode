```python
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        right = 0
        left = 0
        valid = 0
        res = []
        window = {}
        target = {}

        #target

        for c in p: 
            target[c] = target.get(c, 0)+1  

        while right < len(s):
            c = s[right]
            right += 1
            
            #ctarget
            if c in target:
                window[c] = window.get(c,0)+1
                if window[c] == target[c]:
                    valid += 1

            while left < right and right - left >= len(p):
                if valid == len(target) and right-left == len(p):
                    res.append(left)
                d = s[left]
                if d not in target:
                    left +=1
                elif window[d] <= target[d]:
                    break
                else:
                    window[d] -=1
                    left +=1
            # print("window record: "+str(window))
            # print("window content: "+str(s[left:right]))
            # print("valid: "+str(valid))
            # print("res: "+str(res))
        return res

```
