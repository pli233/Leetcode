```python
class Solution:
    def shipWithinDays(self, weights: List[int], days: int) -> int:

        #minimum should be the max val in the weights since thats the single 
        #shipment we need to do for one time
        left = max(weights)

        #maximum should be the sum of all weights, 
        #since we can ship them in one day
        right = sum(weights)+1

        #use binary search to find the least capacity
        #that can ship packages within days day

        while left < right:
            mid = left + (right-left) // 2
            val = self.fx(weights = weights, capacity = mid)

            #尝试搜索左边半区
            if val == days:
                right = mid
            
            #还没达到允许的最大days，说明这个capacity可能大了，搜索左边半区
            elif val < days:
                right = mid
            #超过最大值，增大capacity 搜索右边半区，
            elif val > days:
                left = mid+1
        return left

    
    #define function that given the capacity, 
    #return the number of days we need to ship
    def fx(self, weights: List[int], capacity: int) -> int:

        days = 1
        cursum = 0
        for w in weights:
            cursum += w
            if cursum > capacity:
                days += 1
                cursum = w
        return days
```
