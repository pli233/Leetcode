```python
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        
        left = 0
        right = len(nums)

        while left < right:
            mid = left + (right - left) // 2
            val = nums[mid]

            if val == target:
                return mid
            elif val < target:
                left = mid + 1
            elif val > target:
                right = mid
        return left
```
