```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        ps = 0
        pt = 0
        matched = 0

        while ps < len(s):
            vs = s[ps]

            while pt < len(t):
                vt = t[pt]
                if vs == vt:
                    matched += 1
                    pt += 1
                    break
                pt += 1
            ps += 1
        return matched == len(s)
        
        
```
