```python
class Solution:
    def splitArray(self, nums: List[int], k: int) -> int:
        
        #min number of largestsum should be the max of the array
        #so the max can have its own array
        left = max(nums)

        #max number of largestsum should be the sum of the array
        #so that we only need one array which is itself
        right = sum(nums)+1

        while left < right:
            mid = left + (right - left) // 2
            val = self.fx(nums = nums, largestsum = mid)

            #val等于k，找到target尝试搜索左边半区，如果没有update的话就会返回left = right = cur mid
            if val == k:
                right = mid
            
            #val小于k，说明我们还可以尝试增大val的值，搜索左边半区
            elif val < k:
                right = mid
            
            #val 大于k，说明我们要减小fx的output，所以增大largestsum的input，所以搜索右边半区
            elif val > k:
                #mid一定不符合条件，所以搜索剩下的
                left = mid + 1
        return left

    #define a function give the largest sum in subarray
    #return the number of k subarray it need to form the nums

    def fx(self, nums: List[int], largestsum: int) -> int:
        
        cursum = 0
        nums_array = 1
        for i in nums:  
            cursum += i
            if cursum > largestsum:
                nums_array += 1
                cursum = i
        return nums_array
```
