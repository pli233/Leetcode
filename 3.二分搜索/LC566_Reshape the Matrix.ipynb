```python
class Solution:
    def matrixReshape(self, mat: List[List[int]], r: int, c: int) -> List[List[int]]:
        m, n = len(mat), len(mat[0])
        #  reshape
        if r * c != m * n:
            return mat
        matrix_1d = []

        for row in mat:
            for e in row:
                matrix_1d.append(e)
        res = [[0] * c for _ in range(r)]

        for i in range(r):
            for j in range(c):
                res[i][j] = matrix_1d.pop(0)
        return res
```
