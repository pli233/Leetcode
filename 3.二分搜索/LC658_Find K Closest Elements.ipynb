```python
class Solution:
    def findClosestElements(self, arr: List[int], k: int, x: int) -> List[int]:
        #  x 
        p = self.left_bound(arr, x)

        #  (left, right)
        left = p - 1
        right = p
        res = []

        while len(res) < k:
            # no more val on left
            if left < 0:
                res.append(arr[right])
                right += 1
            # no more val on right
            elif right >= len(arr):
                res.insert(0, arr[left])
                left -= 1
            elif abs(x - arr[left]) > abs(x - arr[right]):
                res.append(arr[right])
                right += 1
            else:
                res.insert(0, arr[left])
                left -= 1
        return res

    # 
    def left_bound(self, nums: List[int], target: int) -> int:
        left = 0
        right = len(nums)

        while left < right:
            mid = left + (right - left) // 2
            if nums[mid] == target:
                right = mid
            elif nums[mid] < target:
                left = mid + 1
            elif nums[mid] > target:
                right = mid
        return left
```
