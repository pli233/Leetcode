```python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        m = len(matrix)
        n = len(matrix[0])

        left = 0
        right = m * n

        while left < right:
            mid  = left + (right- left) // 2
            val = self.get2D(matrix, mid)

            if val == target:
                return True
            #val 小了，搜索右半区
            elif val < target:
                left = mid+1
            #val 大了，搜索左半区
            elif val > target:
                right = mid
        return False

    def get2D(self, matrix: List[List[int]], index: int) -> bool:
        m = len(matrix)
        n = len(matrix[0])

        i = index // n
        j = index % n
        return matrix[i][j]
```
