```python
class Solution:
    def advantageCount(self, nums1: List[int], nums2: List[int]) -> List[int]:

        maxpq = []

        for i in range(len(nums2)):
            heapq.heappush(maxpq, (-nums2[i], i))
        
        left = 0
        right = len(nums1)-1
        nums1.sort()
        res = [0] * len(nums1)
        while maxpq:
            nums2_max,nums2_index = heapq.heappop(maxpq)
            nums2_max *= -1
            nums1_max = nums1[right]

            if nums1_max > nums2_max:
                res[nums2_index] = nums1_max
                right -= 1
            else:
                res[nums2_index] = nums1[left]
                left += 1
        return res
```
