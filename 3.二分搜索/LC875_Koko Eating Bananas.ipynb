```python
class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        #首先我们应该吃尽量多的banna

        left = 1
        right = max(piles)+1


        while left < right:
            mid = left + (right - left) // 2
            val = self.f(piles, mid)
            if val == h:
                # 搜索左侧边界，则需要收缩右侧边界
                right = mid
            elif val < h:
                # 需要让 f(x) 的返回值大一些
                right = mid
            elif val > h:
                # 需要让 f(x) 的返回值小一些
                left = mid+1
        return left 
        

    # 定义：速度为 x 时，需要 f(x) 小时吃完所有香蕉
    # f(x) 随着 x 的增加单调递减
    def f(self, piles: List[int], x: int) -> int:
        hours = 0
        for i in range(len(piles)):
            hours += piles[i] // x
            if piles[i] % x > 0:
                hours += 1
        return hours
```
