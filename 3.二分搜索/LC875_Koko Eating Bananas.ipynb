```python
class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        #banna

        left = 1
        right = max(piles)+1


        while left < right:
            mid = left + (right - left) // 2
            val = self.f(piles, mid)
            if val == h:
                # 
                right = mid
            elif val < h:
                #  f(x) 
                right = mid
            elif val > h:
                #  f(x) 
                left = mid+1
        return left 
        

    #  x  f(x) 
    # f(x)  x 
    def f(self, piles: List[int], x: int) -> int:
        hours = 0
        for i in range(len(piles)):
            hours += piles[i] // x
            if piles[i] % x > 0:
                hours += 1
        return hours
```
