```python
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:

        #Define dp table, dp[n] ncoincoin
        # Initialize dp table with -1, size n+1
        dp = [float("inf")] * (amount+1)
        dp[0] = 0
        for t in range(1, amount+1):
            min_step = float("inf")
            for coin in coins:
                if t-coin >=0:
                    min_step = min(min_step, dp[t-coin]+1)
            dp[t] = min_step
        if dp[amount] == float("inf"):
            return -1
        return dp[amount]
```
