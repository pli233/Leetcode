```python
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:

        # DP,targetinitializetarget+10,solutiondp[target]
        # dp[i]targeti
        # bellman equation: dp[i] = for val in candidates: dp[i-val]+dp[i]
        dp = [[] for _ in range(target + 1)]
        dp[0] = [[]]  #  0 

        for val in candidates:
            for i in range(val, target + 1):
                for comb in dp[i - val]:
                    dp[i].append(comb + [val])

        return dp[target]
```
