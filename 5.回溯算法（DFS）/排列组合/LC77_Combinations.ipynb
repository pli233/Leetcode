```python
class Solution:
    def __init__(self):
        self.res = []
        self.layer = 0
        self.path = []

    def combine(self, n: int, k: int) -> List[List[int]]:

        # range 1-n
        # k is the number of integers in the subset
        # which is the depth of the tree
        self.dfs(1, n, k)
        return self.res

    def dfs(self, start: int, end: int, depth: int):

        for i in range(start, end+1):
            self.path.append(i)
            if len(self.path) == depth:
                self.res.append(self.path.copy())
            self.dfs(i+1, end, depth)
            self.path.pop()
```
