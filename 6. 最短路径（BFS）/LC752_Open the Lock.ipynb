```python
class Solution:
    def openLock(self, deadends: List[str], target: str) -> int:
        if "0000" in deadends:
            return -1
            
        s = "0000"
        if s == target:
            return 0
        
        visited = set()
        visited.add(s)

        q = deque()
        q.append(s)

        step = 0
        while q:
            sz = len(q)
            #
            for i in range(sz):
                cur = q.popleft()

                if cur == target:
                    return step
                
                for neighbor in self.getNeighbors(cur):
                    if neighbor in visited or neighbor in deadends:
                        continue
                    visited.add(neighbor)
                    q.append(neighbor)
            step += 1
        return -1
    
    def getNeighbors(self, s: str):
        possible_slots = [
            [1, 9], #0
            [0, 2], #1
            [1, 3], #2
            [2, 4], #3
            [3, 5], #4
            [4, 6], #5
            [5, 7], #6
            [6, 8], #7
            [7, 9], #8
            [8, 0], #9

        ]
        neighbors = []
        # Iterate over each position in the input string
        for i in range(len(s)):
            current_digit = int(s[i])  # Get the current digit
            
            # Roll the current digit to all possible values
            for new_digit in possible_slots[current_digit]:
                # Replace the digit at position i and form a new string
                new_state = s[:i] + str(new_digit) + s[i + 1:]
                neighbors.append(new_state)
        
        return neighbors
```
