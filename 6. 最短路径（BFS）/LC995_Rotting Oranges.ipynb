```python
class Solution:
    def orangesRotting(self, grid: List[List[int]]) -> int:
        rep = self.map2D(grid)
        q = collections.deque([])
        m = len(grid[0])
        # freshque
        count = 0 
        for i in range(len(rep)):
            if rep[i] == 2:
                q.append(i)
            if rep[i] == 1:
                count += 1

        if count == 0:
            return 0
        step = 0
        while q:
            print(f"current que: {q}")
            print(f"current rep: {rep}")
            sz = len(q)
            step += 1
            for i in range(sz):
                index = q.popleft()
                print(f"{index} start to propogate")
                up = index - m
                if up >= 0 and rep[up] == 1:
                    rep[up] = 2
                    q.append(up)

                down = index + m
                print(f"down index is : {down}")
                if down < len(rep) and rep[down] == 1:

                    rep[down] = 2
                    q.append(down)

                left = index - 1
                if index % m != 0 and rep[left] == 1:
                    rep[left] = 2
                    q.append(left)

                right = index + 1
                if index % m != m - 1 and rep[right] == 1:
                    rep[right] = 2
                    q.append(right)
        
        for val in rep:
            if val == 1:
                return -1
        return step-1

    def map2D(self, grid: List[List[int]]):
        m, n = len(grid), len(grid[0])
        res = []

        for i in range(m):
            for j in range(n):
                res.append(grid[i][j])
        return res
```
