```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def __init__(self):
        self.res = 0

    def sumRootToLeaf(self, root: Optional[TreeNode]) -> int:
        self.traverse(root, [])
        return self.res

    def traverse(self, root: Optional[TreeNode], path: List[int]):
        if root is None:
            return None

        path.append(root.val)
        if root.left is None and root.right is None:
            self.res += self.convert(path)
        self.traverse(root.right, path)
        self.traverse(root.left, path)
        path.pop()

    def convert(self, path: List[int]) -> int:

        res = 0
        p = len(path) - 1
        i = 0

        while p >= 0:
            res += path[p] * (2**i)
            p -= 1
            i += 1
        return res
```
