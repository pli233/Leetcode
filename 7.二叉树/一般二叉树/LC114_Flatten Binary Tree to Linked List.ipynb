```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def flatten(self, root: Optional[TreeNode]) -> None:
        """
        Do not return anything, modify root in-place instead.
        """

        if root is None:
            return

        #flatten the left and right child tree of the current root
        self.flatten(root.left)
        self.flatten(root.right)
        
        #record the left and right
        left = root.left
        right = root.right

        #attach left tree as right child
        root.left = None
        root.right = left

        #go to the end of the right child
        p = root
        while p.right:
            p = p.right
        #assign the right as the child of the right end
        p.right = right
        
```
