```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def __init__(self):
        self.count = [0] * 10
        self.res = 0

    def pseudoPalindromicPaths (self, root: Optional[TreeNode]) -> int:
        self.traverse(root, [])
        return self.res
    
    def traverse(self, root: Optional[TreeNode], path: List[int]):
        if root is None:
            return None
        
        self.count[root.val] += 1
        if root.left is None and root.right is None:
            if self.isPalindrome(self.count):
                self.res += 1
        self.traverse(root.left, path)
        self.traverse(root.right, path)
        self.count[root.val] -= 1


    def isPalindrome(self, count: List[int]) -> bool:

        odd = 0
        for val in count:
            if val % 2 == 1:
                odd += 1
        if odd > 1:
            return False
        return True
```
