```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:

    def findDistance(self, root: Optional[TreeNode], p: int, q: int) -> int:
        if p == q:
            return 0
        # 找到 p 和 q 的路径
        p_path = self.traverse(root, p, [])
        q_path = self.traverse(root, q, [])
        
        # 找到路径分叉点
        i = 0
        while i < len(p_path) and i < len(q_path) and p_path[i] == q_path[i]:
            i += 1

        # 距离为两个路径减去公共路径长度
        return (len(p_path) - i) + (len(q_path) - i)

    def traverse(self, root: Optional[TreeNode], target: int, path: List[int]) -> List[int]:
        if root is None:
            return None

        path.append(root.val)

        # 如果找到目标节点，返回路径one
        if root.val == target:
            return path.copy()

        # 递归地搜索左子树
        left_result = self.traverse(root.left, target, path)
        if left_result:  # 如果左子树找到了路径
            return left_result

        # 递归地搜索右子树
        right_result = self.traverse(root.right, target, path)
        if right_result:  # 如果右子树找到了路径
            return right_result

        # 回溯
        path.pop()

        # 如果没有找到路径，返回空列表
        return None
```
