```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.

        :type root: TreeNode
        :rtype: str
        """
        if root is None:
            return "#"

        left = self.serialize(root.left)
        right = self.serialize(root.right)
        subtree = f"{root.val},{left},{right}"
        return subtree

    def deserialize(self, data):
        """Decodes your encoded data to tree.

        :type data: str
        :rtype: TreeNode
        """
        data = [x for x in data.split(",") if x]
        return self.traverse(data)
        
    def traverse(self, preorder):
        if not preorder:
            return None

        first = preorder.pop(0)
        if first == "#":
            return None
        root = TreeNode(int(first))
        root.left = self.traverse(preorder)
        root.right = self.traverse(preorder)

        return root

# Your Codec object will be instantiated and called as such:
# ser = Codec()
# deser = Codec()
# ans = deser.deserialize(ser.serialize(root))
```
