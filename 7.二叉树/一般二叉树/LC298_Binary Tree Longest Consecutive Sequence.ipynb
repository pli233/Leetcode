```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def __init__(self):
        self.maxLength = 1

    def longestConsecutive(self, root: Optional[TreeNode]) -> int:
        self.dfs(root, 1)
        return self.maxLength

    def dfs(self, root: Optional[TreeNode], pathLength: int):

        if root is None:
            return

        if root.left:
            if root.left.val == root.val + 1:
                self.maxLength = max(self.maxLength, pathLength + 1)
                self.dfs(root.left, pathLength + 1)
            else:
                self.dfs(root.left, 1)

        if root.right:
            if root.right.val == root.val + 1:
                self.maxLength = max(self.maxLength, pathLength + 1)
                self.dfs(root.right, pathLength + 1)
            else:
                self.dfs(root.right, 1)
```
