```python
class Solution:
    def numTrees(self, n: int) -> int:
        #  [1, n]  BST 
        return self.count(1, n)

    #  [lo, hi]  BST 
    def count(self, lo: int, hi: int) -> int:
        # base case, when there is no more subtree, we return 1
        if lo > hi: 
            return 1

        res = 0
        for i in range(lo, hi+1):
            # i  root
            # i = 1, low = 1, i-1 = 0 , left = 1
            left = self.count(lo, i - 1)
            right = self.count(i + 1, hi)
            #  BST 
            res += left * right

        return res
```
