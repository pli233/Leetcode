```python
import heapq
class Solution:
    def kthSmallest(self, matrix: List[List[int]], k: int) -> int:

        pq = []
        for row in matrix:
            for e in row:
                heapq.heappush(pq, e)
        
        res = -1
        for _ in range(k):
            res = heapq.heappop(pq)
        return res
```
